```python
import pandas as pd  

# Load the dataset
main_data = pd.read_csv("main_data.csv")

# Clean column names by stripping extra whitespace and line breaks
main_data.columns = main_data.columns.str.strip().str.replace("\r\n", " ", regex=True)

# Define the correct desired column order
desired_columns = [
    "STATE", "CONSTITUENCY", "NAME", "WINNER", "PARTY", "SYMBOL", "GENDER", 
    "CRIMINAL CASES", "AGE", "CATEGORY", "EDUCATION", "ASSETS", "LIABILITIES", 
    "GENERAL VOTES", "POSTAL VOTES", "TOTAL VOTES", 
    "OVER TOTAL ELECTORS IN CONSTITUENCY", "OVER TOTAL VOTES POLLED IN CONSTITUENCY", "TOTAL ELECTORS"
]

# Identify any missing columns from desired_columns
missing_columns = [col for col in desired_columns if col not in main_data.columns]

if missing_columns:
    print(f"⚠️ Warning: These columns are missing in main_data: {missing_columns}")

# Find extra columns that are not in the desired order
extra_columns = [col for col in main_data.columns if col not in desired_columns]

# Create final column order (existing desired columns first, then extra columns)
final_column_order = [col for col in desired_columns if col in main_data.columns] + extra_columns

# Reorder the DataFrame
main_data = main_data[final_column_order]

# Save the updated dataset
main_data.to_csv("main_data.csv", index=False)

print("✅ Columns rearranged successfully!")
```
